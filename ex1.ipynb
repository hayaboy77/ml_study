```python
def maxi(li):
    maximum=-99999
    for i in li:
        if maximum < i:
            maximum =i 
    return maximum
```

```python
li=[273,52,103,57,271]
```

```python
result=maxi(li)
```

```python
result
```

```python
def max_list(a):
    j = 0
    for i in a:
        if i > j :
            j=i
    return j
```

```python
with open('example.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)          # ① writer 함수 사용
    writer.writerow(['이름', '나이', '도시'])  # ② writerow 메서드 사용
    writer.writerow(['홍길동', 25, '서울'])
    writer.writerow(['김철수', 30, '부산'])
    # csvfile.close()는 with 문 덕분에 자동으로 호출됨 (④ close는 파일 객체 메서드)
```
