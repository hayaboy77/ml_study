```python
train_input, test_input, train_target, test_target = train_test_split(fish_data,fish_target,stratify=fish_target,  random_state=42)
```

```python
len(train_target[train_target==1]), len(train_target[train_target==0])
```

```python
# 정규화 하지 않고(척도(scale), 길이, 무게가 다른 단위) 할 경우, 단점(학습 시간이 오래걸리고, 정확도가 낮아짐)
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

```python
# 길이 25cm이고 무게가 150인 도미인지, 빙어인지 모르는 어떤 생선을 예측
```

```python
print(kn.predict([[25, 150]])) #빙어라고 잘 못 예측함
```
